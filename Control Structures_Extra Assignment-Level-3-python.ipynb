```python
#Catagory the people by their age like children, adult, citizen, senior citizen...
#age<18-->children
#age<35-->adult
#age<59-->citizen
#age>59-->senior citizen

age=int(input("Enter your age: "))
if(age<18):
    print("Children")
elif(age<35):
    print("Adult")
elif(age<59):
    print("Citizen")
else:
    print("Senior Citizen")
```
