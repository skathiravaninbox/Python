```python
# print Odd Numbers in the list
List=[14,88,24,19,20,77]
for num in List:
    if(num%2!=0):
        print(num," is Odd")
```

```python
# print Even Numbers in the list
List=[14,88,24,19,20,77]
for num in List:
    if(num%2==0):
        print(num," is Even")
```
