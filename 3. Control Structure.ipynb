```python
List=[13,81,34,55,90,23,15]
for temp in List:
    print(temp)
```

```python
#for and if --> multiple inputs checked in a set of procedure
for age in List:
    if(age<18):
        print("Children")
    elif(age<35):
        print("Adult")
    elif(age<59):
        print("Citizen")
    else:
        print("Senior Citizen")
```
